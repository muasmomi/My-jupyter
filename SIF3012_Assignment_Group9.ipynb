To solve this, we will be using a useful numerical approach that we have been studied throughout this semester and the method is
called shooting method.

1) Drag force equation:

$$ F_{drag}= \frac{1}{2}C_dAv^2$$
where 
$A$= Reference area,
 $$= density,
 $Cd$= drag coeficient,
 $v$= flow velocity relative to object

2) Rolling friction equation:

$$ F_r= _rNcos() $$
where
$_r$= coeficient of rolling friction,
$N$= Normal force

3) Gravity equation:

$$ F_{gravity}= mgsin() $$

Using Newtonian mechanics concept 

$$F=ma$$
$$ma=-F_{drag} - F_r + F_{gravity}$$
$$ma=-\frac{1}{2}C_dAv^2 - _rNcos() + mgsin()$$
$$a=-\frac{0.5C_dAv^2}{m} - _rgcos() + gsin()$$
