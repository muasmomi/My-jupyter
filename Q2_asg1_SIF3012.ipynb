```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy.optimize import minimize
y=[1,-1.4,-7.5,-12.4,0]
x=[1,2,3,4,5]
plt.scatter(x,y)

```

```python
from scipy.interpolate import interp1d
f= interp1d(x,y,kind='cubic')
x_dense= np.linspace(1,5,50)
y_dense= f(x_dense)
plt.plot(x_dense,y_dense)
```
